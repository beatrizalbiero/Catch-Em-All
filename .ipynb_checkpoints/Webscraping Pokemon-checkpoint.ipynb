### Libs

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import StaleElementReferenceException
import csv
from bs4 import BeautifulSoup as bs
import re
import pprint as pp
```

### Home

```python
browser = webdriver.Chrome(executable_path=r"chromedriver.exe")
url = 'http://www.serebii.net/'
html = browser.page_source
browser.get(url)
browser.implicitly_wait(2)
```

### Catch First Pokemon

```python
browser.find_element_by_xpath('//*[@id="menu"]/a[9]').click()
menu = Select(browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[3]/table/tbody/tr[1]/td[1]/form/select'))
menu.select_by_index('1')
```

### Creates CSV File

```python
with open ('pokemons.csv', 'w') as y:
    y.write('Name;Classification;Height;Weight;Capture Rate;Type\n')  
```

### Gotta Catch 'Em All

```python
#getting a list of ranges
html = browser.page_source
test = bs(html,'html.parser')
test = str(test)
matches = re.findall('[A-Z][a-z]+: [0-9]{3} ?- ?[0-9]{3}',test)
matches = list(map(lambda x: x[-3:], matches))
for j in range(1,8):
    menu = Select(browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/table/tbody/tr[1]/td[' +
                                                str(j) + ']/form/select'))
    for i in range(1,int(matches[0]) + 1):
        html = browser.page_source
        menu = Select(browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/table/tbody/tr[1]/td[' +
                                                    str(j) + ']/form/select'))
        menu.select_by_index(i)
        name = browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/div/table[1]/tbody/tr/td[1]/table/tbody/tr/td[2]/font/b').text
        name = re.sub('[^0-9a-zA-Z]+', '*', name) # ignore non alpha numerical characters
        name = re.sub('\#[0-9]*' + ' ','',name) # Select only the Pokemon's name
        classification = browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/div/p[1]/table[1]/tbody/tr[4]/td[1]').text
        height = browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/div/p[1]/table[1]/tbody/tr[4]/td[2]').text.split('\n')[1]
        weight = browser.find_element_by_xpath('/html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/div/p[1]/table[1]/tbody/tr[4]/td[3]').text.split('\n')[1]
        capture = browser.find_element_by_xpath('//html/body/table[2]/tbody/tr[2]/td[2]/font/div[2]/div/p[1]/table[1]/tbody/tr[4]/td[4]').text
        pkm_page = bs(html,'html.parser')
        x = pkm_page.find_all("td", class_="tooltabcon")
        matches = re.findall('(pokedex-sm/[a-z]*\.)',str(x))
        types = ''
        for pat in matches:
            t = pat[pat.find('/')+1:-1]
            types = (types + t + ', ')
        types = types.strip(', ')
        line = str(name + ';' + classification + ';' + height + ';' + weight + ';' + capture + ';' + types + '\n')
        line = '' + line       
        with open ('pokemons.csv', 'r+') as y:
            y.readline()
            y.write(line) 

        pp.pprint(line)
```
